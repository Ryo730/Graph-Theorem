```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
##Visualize f(t)=cos(t)+sin(t)
t=np.arange(0,100)
cost=np.cos(t)
sint=np.sin(t)
f_t=cost+sint
plt.plot(t,f_t)
plt.title("f(t)=cos(t)+sin(t)")
plt.xlabel("time")
plt.ylabel("value")
plt.show()
```

```python
##Visualize f_x,f_y,f_xy
t=np.arange(0,100)
x=np.cos(t)
y=np.sin(t)
##First make 100 random value 
#randint only make integer.So divide by 100
e=np.random.randint(-10,10,(100))/100
f_x=x
f_y=y
f_xy=x+y
#care about e[t].That's a list including 100　elements.
f_xye=x+y+e[t]
fig,axe=plt.subplots(2,2,figsize=(9,5))
#assign figure to the place
axe[0,0].plot(t,f_x)
axe[0,1].plot(t,f_y)
axe[1,0].plot(t,f_xy)
axe[1,1].plot(t,f_xye)
#attach label and title 
axe[0,0].set_title('f(x)')
axe[0,0].set_xlabel("time")
axe[0,0].set_ylabel("value")
axe[0,1].set_title('f(y)')
axe[0,1].set_xlabel("time")
axe[0,1].set_ylabel("value")
axe[1,0].set_title('f(x+y)')
axe[1,0].set_xlabel("time")
axe[1,0].set_ylabel("value")
axe[1,1].set_title('f(x+y+e)')
axe[1,1].set_xlabel("time")
axe[1,1].set_ylabel("value")
#For not　to  overlap figures
plt.tight_layout()
plt.show()
```
