```python
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

１．ノード数３の一方向リンクグラフ

```python
##空の有向グラフ空間を作成
G_1=nx.DiGraph()
```

```python
##ノードを追加
for i in range(1,4):
    G_1.add_node(i)
```

```python
## edge(リンク)を生成
G_1.add_edge(1,2)
G_1.add_edge(2,3)
```

```python
nx.draw_networkx(G_1)
plt.show()
```

ノード数５の完全グラフ

```python
##完全グラフの作成
#ノード番号が０からだと気持ち悪いので１からに調整した
G_2=nx.complete_graph(6)
G_2.remove_node(0)
```

```python
nx.draw_networkx(G_2)
plt.show()
```

３．ノード数７のパスグラフの作成

```python
G_3=nx.path_graph(8)
G_3.remove_node(0)
```

```python
nx.draw_networkx(G_3)
plt.show()
```

5．重み付き有向グラフ　リンク本数７
（リンク情報のファイルを作成のちグラフ作成）

```python
##空の有向グラフ空間を作成
G_5=nx.DiGraph()
```

```python
file=open('hoge.txt')
print(file.read())
```

```python
G_5= nx.read_weighted_edgelist('hoge.txt', nodetype=int, create_using=nx.DiGraph)
```

```python
pos = nx.spring_layout(G_5)
edge_labels = {(i, j): w['weight'] for i, j, w in G_5.edges(data=True)}
```

```python
nx.draw_networkx_edge_labels(G_5,pos,edge_labels=edge_labels)
nx.draw_networkx(G_5,pos,with_labels=True)
plt.show()
```
