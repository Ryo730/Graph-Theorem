```python
def describe(la):
    w=np.array(nx.to_numpy_matrix(la))
    l=-w
    for i in range(0,len(w)):
        l[i][i]=w[i].sum()

    eig,vector=np.linalg.eig(l)
    vec=vector.T
    print("eig：{}".format(eig))
    print("vector：{}".format(vec))
    deg=list(nx.degree(la))
    print("degree：{}".format(deg))
    deg_w=list(la.degree(weight='weight'))
    print("weighted_degree：{}".format(deg_w))
    nx.draw_networkx(la)
    plt.show()
    
##visualize fiedler-vector
def vis_fie(la):
    w=np.array(nx.to_numpy_matrix(la))
    l=-w
    for i in range(0,len(w)):
        l[i][i]=w[i].sum()

    eig,vector=np.linalg.eig(l)
    vec=vector.T
    data=pd.DataFrame(data=[eig,vec]).T
    data=data.sort_values(0)
    data=data.reset_index(drop=True)
    
    
    i=0
    while np.around(data[0][i],decimals=10)!=0:
        i+=1
    ##If u dont,tiny value exists　instead of 0 
    fiedler=data[1][i+1]
    x=data.index+1
    y=fiedler
    plt.xlabel("node ID")
    plt.scatter(x,y)
    print('第二最小固有値：{}'.format(data[0][i+1]))

    
g_list=[G_1,G_2,G_3,G_4,G_5]
for i in g_list:
    describe(i)
    vis_fie(i)
    plt.show()
```
